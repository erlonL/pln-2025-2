```python
def get_next_word_with_most_prob(bigram_counts, w1):
    # get all bigrams that start with w1
    new_bigrams = [bgc for bgc in bigram_counts if bgc[0] == w1]
    most_probable = Counter(new_bigrams).most_common(1)
    return most_probable[0][0][1]

random_word = random.choice(words)

next_word = random_word

phrase_list = []
for i in range(5):
    phrase_list.append(next_word)

    next_word = get_next_word_with_most_prob(bigram_counts, next_word)

" ".join(phrase_list)
```

```python
def join_words(word_list):
    # if it is a [. ! ?], the next word will be capitalized and dont add a space
    # if it is a word, append with space normally
    CAP_PONCTS = [".", "!", "?"]
    NORMAL_PONCTS = [",", "\'", '\"', ";"]
    DASH_PONCT = "-"

    # transform word list to indexed dict
    word_set = {i: word for i, word in enumerate(word_list)}

    result_string = ""
    for i, word in list(word_set.items()):
        if i == 0:
            result_string += f"{word.capitalize()}"
            continue

        if word in CAP_PONCTS:
            result_string += f"{word}"
            word_set[i+1] = word.capitalize()
        elif word in NORMAL_PONCTS:
            result_string += f"{word}"
        else:
            result_string += f" {word}"

    return result_string
```

```python
def get_next_word_with_random_choice_top10(bigram_counts, w1):
    # get all bigrams that start with w1
    new_bigrams = [bgc for bgc in bigram_counts if bgc[0] == w1]
    most_probables = Counter(new_bigrams).most_common(10)
    
    choice = random.choice(list(most_probables))[0][1]

    return choice
```

```python
def generate_words_with_stop(stop_word, choice_func, max_iters=100):
    global words, bigram_counts
    random_word = random.choice(words)

    next_word = random_word

    STOP_ITERS = max_iters
    word_index = 0

    phrase_as_list = []
    while next_word != stop_word:
        if word_index >= STOP_ITERS:
            break
        phrase_as_list.append(next_word)
        next_word = choice_func(bigram_counts, next_word)
        
        word_index += 1
    phrase_as_list.append(next_word)

    return join_words(phrase_as_list)
```

```python
generate_words_with_stop(".", get_next_word_with_random_choice_top10)
```

```python
def get_next_word_with_least_prob(bigram_counts, w1):
    # get all bigrams that start with w1
    new_bigrams = [bgc for bgc in bigram_counts if bgc[0] == w1]
    least_prob = Counter(new_bigrams).most_common()[-1]

    return least_prob[0][1]
```

```python
generate_words_with_stop(".", get_next_word_with_least_prob, 10)
```

```python
def get_next_word_randomly(bigram_counts, w1):
    # get all bigrams that start with w1
    new_bigrams = [bgc for bgc in bigram_counts if bgc[0] == w1]
    random_choice = random.choice(new_bigrams)

    return random_choice[1]
```
