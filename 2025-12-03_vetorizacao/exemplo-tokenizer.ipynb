```python
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer()
word_tokenizer = count_vectorizer.build_tokenizer()

word_tokenizer("O meu relógio conta as horas melhor do que o seu.")
```

```python
import nltk

stopwords = nltk.corpus.stopwords.words('portuguese')
stopwords[:10], len(stopwords)
```

```python
count_vectorizer = CountVectorizer(stop_words=stopwords)
analyzer = count_vectorizer.build_analyzer()

analyzer("O meu relógio conta as horas melhor do que o seu.")
```

```python
stemmer = nltk.stem.RSLPStemmer()

text = "joão era o melhor jogador do campeonato. ele ganhou diversos prêmios"
[stemmer.stem(word) for word in word_tokenizer(text)]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

documents = [
    "O rato roeu a roupa do rei de Roma. A roupa ficou rasgada",
    "A rainha raivosa rasgou o resto da roupa",
    "O rato roeu a roupa da rainha raivosa"
]

# One Hot Encoding
count_vectorizer_ohe = CountVectorizer(binary=True, stop_words=stopwords)
X_ohe = count_vectorizer_ohe.fit_transform(documents)

# Bag of Words (para comparação)
count_vectorizer_bow = CountVectorizer(stop_words=stopwords)
X_bow = count_vectorizer_bow.fit_transform(documents)

import pandas as pd
print("One Hot Encoding:")
print(pd.DataFrame(X_ohe.toarray(), columns=count_vectorizer_ohe.get_feature_names_out()))
print("\nBag of Words:")
print(pd.DataFrame(X_bow.toarray(), columns=count_vectorizer_bow.get_feature_names_out()))
```

```python
# bag of words
documents = [
    "O rato roeu a roupa do rei de Roma",
    "A rainha de Roma ficou muito brava com o rato",
    "O rato fugiu para a floresta"
]

count_vectorizer = CountVectorizer(stop_words=stopwords)
X = count_vectorizer.fit_transform(documents)
X.toarray(), count_vectorizer.get_feature_names_out()

# plot a table with pandas
import pandas as pd

df = pd.DataFrame(X.toarray(), columns=count_vectorizer.get_feature_names_out())
df
```
