Como encontrar um string que traz uma lista de nomes de pessoas da 
seguinte forma: 

‚Ä¢ ana;pedro;maria;lucas;

```python
import re
```

```python
text = 'ana;pedro;maria;lucas'
regex = r'[^;]*'

re.findall(regex, text)
```

```python
text = 'ana;pedro;maria;lucas'
regex = r'(+;)+'

re.findall(regex, text)
```

```python
text = '''
78012-230 23519235 asdasd qwed asd 12312513 1231256 -12312454 123156-341 12315-123512
12398751 3452-1315 -12315-123 5123-51324 
'''
regex = r'\b[0-9]{5}-?[0-9]{3}\b'

re.findall(regex, text)
```

Vamos fazer uma valida√ß√£o para e-mails com os seguintes requisitos:   
‚Ä¢ Um email √© composto de usu√°rio@dom√≠nio  

‚Ä¢ Usu√°rio √© uma sequ√™ncia com um ou mais caracteres que podem ser:  
- letras, d√≠gitos ou underline

‚Ä¢ Dom√≠nio √© uma sequ√™ncia com um ou mais caracteres que podem ser:
- letras, d√≠gitos ou pontos.

‚Ä¢ Ele precisa terminar com ‚Äú.X‚Äù, onde X √© uma sequ√™ncia de no m√≠nimo 2 letras

```python
email = "yuri@ci.ufpb.br"

regex = r'([a-zA-Z]+)@([a-zA-Z0-9.]+)'

re.findall(regex, email)
```
